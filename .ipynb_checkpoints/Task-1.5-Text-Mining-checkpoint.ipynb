### 2. Importing libraries

```python
from textblob import TextBlob 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

### 3. Tokenization

```python
myfile = open('20th_Century_Wiki.txt', encoding='utf-8')
```

```python
# Import txt file

with open('20th_Century_Wiki.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)

# Limit to the first 20 words

tokenized_word_limited = tokenized_word[:20]
print(tokenized_word_limited)
```

```python
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words) 
```

```python
most_common_words = dist_words.most_common(10)
```

```python
# Separate the words and their frequencies

words = [word for word, freq in most_common_words]
frequencies = [freq for word, freq in most_common_words]
```

```python
# Create the bar chart
plt.figure(figsize=(10, 5))
plt.bar(words, frequencies, color='blue')
plt.xlabel('Words')
plt.ylabel('Frequencies')
plt.title('Top 10 Most Common Words')
plt.xticks(rotation=45)
plt.show()
```

### 4. Stopwords

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
# Create a new FreqDist for filtered_words

dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
dist_words_filter
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
# Tokenize the text
tokenized_word_2 = word_tokenize(sans_punc)

# Limit to the first 20 words
tokenized_word_2_limited = tokenized_word_2[:20]

print(tokenized_word_2_limited)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
most_common_filtered_words = dist_words_filter_2.most_common(20)
```

```python
# Separate the words and their frequencies

words = [word for word, freq in most_common_filtered_words]
frequencies = [freq for word, freq in most_common_filtered_words]
```

```python
# Create the bar chart
plt.figure(figsize=(10, 5))
plt.bar(words, frequencies, color='blue')
plt.xlabel('Words')
plt.ylabel('Frequencies')
plt.title('Top 10 Most Common Words')
plt.xticks(rotation=45)
plt.show()
```

### The results look much better as there are useful words in it. However, there are also still some letters in this top 10 that will need to be filtered out. 

### 5. POS Tags with TextBlob

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The', 'www', 'th']
```

```python
filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
# Limit text to show
text_limited = text[:20]

text_limited
```

```python
tags_list = text.tags
```

```python
# Limit tags to show
tags_list_limited = tags_list[:40]

tags_list_limited
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
```

```python
df_text.head()
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
df_t.head()
```

```python
top10 = df_t.nlargest(10, 'Words')
```

```python
top10
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top10).set_title("20th Century - top 10 word types used")
```

#### 01 Nouns

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15_nouns = y.nlargest(15, 'Occurences')
```

```python
top15_nouns
```

```python
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15_nouns).set_title("20th Century - most frequently used nouns")
```

## Both war and world are mentioned twice as both are written in two different ways. This shows these words are most common by far. Alsmost all words are clearly related to the second world war.

#### 02 Verbs

```python
df = df_text[(df_text['Word type'] == "VB")  | (df_text['Word type'] == "VBD")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15_verbs = y.nlargest(15, 'Occurences')
```

```python
top15_verbs
```

```python
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15_verbs).set_title("20th Century - most frequently used verbs")
```

## Also the verbs are clearly related to the war.

#### 03 Adjectives

```python
df = df_text[df_text['Word type'] == "JJ"]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15_adj = y.nlargest(15, 'Occurences')
```

```python
top15_adj
```

```python
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15_adj).set_title("20th Century - most frequently used adjectives")
```

## The 15 most used adjectives show many country related ones. From which for sure a couple war related.

### 6. Counting country occurences

```python
listToStr = ' '.join([str(elem) for elem in filtered])

#limit on showing list
listToStr_limited = listToStr[:40]
print(listToStr_limited)
```

```python
# Create a count for the main characters

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
all_counts
```

```python
countries = pd.read_csv("countries_list_20th_century_1.5.csv", index_col = 0)
countries.reset_index(drop=True, inplace=True)
countries_list = [country.strip().lower() for country in countries['country_name']]
```

```python
all_counts_lower = {k.lower(): v for k, v in all_counts.items()}
```

```python
dict_of_counts = {d: all_counts_lower[d] for d in countries_list if d in all_counts_lower}
```

```python
dict_of_counts
```

```python
# Search for the names from the list in the dictionary

dct = {v:[k] for v,k in dict_of_counts.items()}  
df = pd.DataFrame(dct)
```

```python
df
```

```python
df = df.transpose().reset_index()
```

```python
# checking the datatype
df.dtypes
```

```python
df
```

```python
df.rename(columns = {"index":"Country", 0:"Times mentioned"}, inplace = True)
```

```python
df
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 20))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x = "Times mentioned", y = "Country",
    saturation=0.9, data = df.sort_values("Times mentioned", ascending = False)).set_title("20th Century - most frequently mentioned countries")
```

## The countries that are the most times mentioned are the countries clearly related to wars during the 20th century
